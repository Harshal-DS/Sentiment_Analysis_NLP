## Task 4

```python
#4
class EVChoice:
    def __init__(self, df):
        self.df = df

    def best_ev(self, min_budget, car_range, battery):
        filtered_df = self.df[
            (self.df["Minimal price (gross) [PLN]"] <= min_budget) &
            (self.df["Range (WLTP) [km]"] >= car_range) &
            (self.df["Battery capacity [kWh]"] >= battery)]
        
        result = filtered_df.sort_values(by="Range (WLTP) [km]", ascending=False).head(3)

        if not result.empty:
            return result[["Car full name", "Minimal price (gross) [PLN]", "Range (WLTP) [km]", "Battery capacity [kWh]"]]
        else:
            print("No EVs match your criteria.")
            
```

```python
ev1 = EVChoice(df)
ev1.best_ev(540000,234,54)
```

### So here in task 5 
### Null Hypothesis(HO):There is no significant difference in the mean engine power between Tesla and Audi models
### Alternate Hypothesis(H1):There is no significant difference in the mean engine power between Tesla and Audi models
### here we cannot reject null hypothesis since p-value is grater than 0.05
### TEST CONCLUSION:There is no significant difference in the average Engine power [KM] of vehicles manufactured by two leading manufacturers i.e. Tesla and Audi.
